# 방문자들의 동선 분석

```python
fig, ax = plt.subplots()

for id in df['id'].unique():
    visitor_df = df[df['id'] == id]
    ax.plot(visitor_df['x'], visitor_df['y'], marker='o')

ax.set_title('Visitors\' paths')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```
